```python
import zipfile
import pandas as pd

zip_file = "data/1000000-bandcamp-sales.zip"  # Replace with the path to your zip file
csv_file = "data/1000000-bandcamp.csv"  # Replace with the name of the CSV file within the zip

# Extract the CSV file from the zip archive
with zipfile.ZipFile(zip_file, "r") as zf:
    # Assuming the CSV file is in the root directory of the zip file
    csv_data = zf.read(csv_file)

# Read the CSV data into a pandas DataFrame
df = pd.read_csv(pd.compat.StringIO(csv_data.decode('utf-8')))

# Now you can work with the DataFrame
print(df.head())
```
